### 교집합, 합집합, 차집합

```python
#리스트로는 &로 합집합 안됨
s1 = [1,2,3,4,5,6,]
s2 = [4,5,6,7,8,9]

s1 & s2
```

```python
s1 = [1,2,3,4,5,6,]
s2 = [4,5,6,7,8,9]

s1 = set(s1)
s2= set(s2)

#교집합 & 

result1 = s1 & s2
print("교집합(&):", result1)
print("교집합(&)[형변환]:", list(result1))

#교집합 intersection()
result2 = s1. intersection(s2)
print("교집합(intersection):", result2)
print("교집합(intersection)[형변환]:", list(result2))
```

```python
s1 = [1,2,3,4,5,6,]
s2 = [4,5,6,7,8,9]

s1 = set(s1)
s2= set(s2)

#합집합 |
result1 = s1 | s2
print("합집합(|):", result1)
print("합집합(|)[형변환]:", list(result1))


# 합집합 union()
result2 = s1.union(s2)
print("합집합(union):", result2)
print("합집합(union)[형변환]:", list(result2))
```

```python
s1 = [1,2,3,4,5,6,]
s2 = [4,5,6,7,8,9]

s1 = set(s1)
s2= set(s2)

#차집합 - 
result1 = s1 - s2
print("차집합(-):", result1)
print("차집합(-)[형변환]:", list(result1))

result2 = s2 - s1
print("차집합(-):", result2)
print("차집합(-)[형변환]:", list(result2))


#차집합 difference ()
result3 = s1.difference(s2)
print("차집합(difference):", result3)
print("차집합(difference)[형변환]:", list(result3))

result4 = s2.difference(s1)
print("차집합(difference):", result4)
print("차집합(difference)[형변환]:", list(result4))
```

### 집합 자료형 관련 함수

```python
# 값 1 개 추가하기
s1 = set ([1,2,3])
s1.add(4)
print(s1)

#2개 요소 추가 시 에러 : add() 하나의 인수만 입력이 가능
#s1.add(4,5)


#리스트 타입의 데이터는 인수로 사용할 수 없음
#s1.add([4])
```

```python
#값 여러개 추가하기 update() - 리스트 형식은 추가 가능
s1 = set ([1,2,3])
s1.update([4,5,6])
print(s1)


#s1.update(4,5) - 이건 에러뜸

```

```python
# 특정값 제거하기 remove() -리스트 형식으로 제거는 불가
s1 = set ([1,2,3])
s1.remove(1)
print(s1)
```
