```python
#csv 파일 쓰기

import csv

with open("./example.csv" ,"w") as file:

    csv_writer = csv.writer(file)
    print(type(csv_writer))

    csv_writer.writerow(["이름","나이","직업"])


    for i in range(10):
        csv_writer.writerow([f"{i}", "홍길동","25","학생"])

    csv_writer.writerow(["홍길동","25","학생"])
    csv_writer.writerow(["길동","25","학생"])
    csv_writer.writerow(["동","25","학생"])


```

```python
# csv 파일 읽기

import csv

with open("./FY2019.csv","r") as file:
    csv_reader = csv.reader(file)

    for row in csv_reader:
        print(row)
        print(type(row))
```
