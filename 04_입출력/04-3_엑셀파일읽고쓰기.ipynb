```python
from openpyxl import Workbook

wb = Workbook()
ws =wb.active

ws.title = "수강생정보"

title = ["번호","이름","과목"]

ws.append(title)

data = [1,"홍길동","수학"]

ws.append(data)

wb.save("./수강생리스트.xlxs")
wb.close()
```
